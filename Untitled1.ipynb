```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pyESN import ESN
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
data = pd.read_csv(io.BytesIO(uploaded['continuous_factory_process.csv']))
```

```python
relevant_columns = [
    'Machine1.RawMaterialFeederParameter.U.Actual',
    'Machine1.Zone1Temperature.C.Actual',
    'Machine1.Zone2Temperature.C.Actual',
    'Stage2.Output.Measurement1.U.Actual',
    'Stage2.Output.Measurement1.U.Setpoint',
    'Stage2.Output.Measurement2.U.Actual',
    'Stage2.Output.Measurement2.U.Setpoint'
]
```

```python
selected_data = data[relevant_columns]
```

```python
sensor_columns = [
    'Machine1.RawMaterialFeederParameter.U.Actual',
    'Machine1.Zone1Temperature.C.Actual',
    'Machine1.Zone2Temperature.C.Actual',
    'Stage2.Output.Measurement1.U.Actual',
    'Stage2.Output.Measurement2.U.Actual'
]
```

```python
normalized_data = (selected_data[sensor_columns] - selected_data[sensor_columns].mean()) / selected_data[sensor_columns].std()
```

```python
setpoints = selected_data[['Stage2.Output.Measurement1.U.Setpoint', 'Stage2.Output.Measurement2.U.Setpoint']]
```

```python
train_input = normalized_data.values[:-1]  # Input is all rows except the last one
train_output = normalized_data.values[1:]  # Output is all rows except the first one
```

```python
esn = ESN(n_inputs=train_input.shape[1], n_outputs=train_output.shape[1], n_reservoir=200, sparsity=0.2, random_state=42)
```

```python
esn.fit(train_input, train_output)
```

```python
predicted_output = esn.predict(train_input)
```

```python
predicted_output_flattened = predicted_output.flatten()
train_output_flattened = train_output.flatten()
```

```python
smoothed_actual = pd.Series(train_output_flattened).rolling(window=5).mean().values
smoothed_predicted = pd.Series(predicted_output_flattened).rolling(window=5).mean().values
```

```python
threshold = 0.5
anomalies = np.where(np.abs(smoothed_predicted - smoothed_actual) > threshold)[0]
```

```python
consecutive_anomalies_threshold = 5
anomaly_groups = np.split(anomalies, np.where(np.diff(anomalies) != 1)[0] + 1)
significant_anomalies = [group for group in anomaly_groups if len(group) > consecutive_anomalies_threshold]
significant_anomalies_flat = np.concatenate(significant_anomalies)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(len(train_output_flattened)), train_output_flattened, label="True Sensor Data")
plt.plot(range(len(predicted_output_flattened)), predicted_output_flattened, label="Predicted Data", linestyle="--")
plt.scatter(significant_anomalies_flat, train_output_flattened[significant_anomalies_flat], color='red', label="Anomalies")
plt.xlabel("Time Step")
plt.ylabel("Sensor Readings")
plt.legend()
plt.title("Anomaly Detection Using ESN (Multiple Sensors, Smoothed)")
plt.show()
```

```python
print(f"Detected {len(anomalies)} anomalies in the data.")
```
